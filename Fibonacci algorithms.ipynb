```python
# Fibonacci function with recursion
fib_cache = {}

def fib(n):
    """Fibonacci function with recursion and cache."""
    if n in fib_cache:
        return fib_cache[n]
    
    if n <= 1:
        return 1

    value = fib(n - 1) + fib(n - 2)
    fib_cache[n] = value
    return value
```

```python
for n in range(100):
    print(n, ':', fib(n))
```

```python
# Fibonacci function without recursion
def fib_dinamic(n):
    fib = [0, 1] + [0] * (n-1)
    for i in range(2, n + 1):
        fib[i] = fib[i-1] + fib[i-2]
    
    return fib
```

```python
fib_dinamic(10)
```
